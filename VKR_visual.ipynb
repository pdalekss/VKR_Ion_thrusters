```python
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline
```

```python
class Particle:
    def __init__(self, type, coords):
        self.type = type
        self.coords = coords
```

```python
def get_random_coords(l, r):
    z = random.uniform(0, l)
    x = random.uniform(-r, r)
    y = random.uniform(-r, r)

    while y > (r**2 - x**2)**(1/2) or -y > (r**2 - x**2)**(1/2):
        y = random.uniform(-r, r)
    return (x, y, z)
```

```python
R = 100
x1 = np.linspace(0, 7, 100)
z1 = np.linspace(-10, 10, 100)
Xc1, Zc1 = np.meshgrid(x1, z1)
Yc1 = np.sqrt(R - Zc1 ** 2)

R1 = 1
x2 = np.linspace(0, 7, 100)
z2 = np.linspace(-1, 1, 100)
Xc2, Zc2 = np.meshgrid(x2, z2)
Yc2 = np.sqrt(R1 - Zc2 ** 2)
```

```python
n = 5000

ions = []
x_ion, y_ion, z_ion = [], [], []

electrons = []
x_el, y_el, z_el = [], [], []
```

```python
for _ in range(n):
    ion = Particle('ion', get_random_coords(7, 10))
    ions.append(ion)
    x_ion.append(ion.coords[0])
    y_ion.append(ion.coords[1])
    z_ion.append(ion.coords[2])

    electron = Particle('electron', get_random_coords(7, 10))
    electrons.append(electron)
    x_el.append(electron.coords[0])
    y_el.append(electron.coords[1])
    z_el.append(electron.coords[2])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_ion, y_ion, z_ion, zdir='x', color='red', s=0.2, label=f'',alpha=0.8)
ax.scatter(x_el, y_el, z_el, zdir='x', color='blue', s=0.2, label=f'',alpha=0.8)

ax.plot_surface(Xc1, Yc1, Zc1, color='black', alpha=0.8)
ax.plot_surface(Xc1, -Yc1, Zc1, color='black', alpha=0.8)

ax.plot_surface(Xc2, Yc2, Zc2, color='black', alpha=1)
ax.plot_surface(Xc2, -Yc2, Zc2, color='black', alpha=1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_box_aspect((100, 100, 100))
frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
frame1.axes.yaxis.set_ticklabels([])
frame1.axes.zaxis.set_ticklabels([])
plt.legend()
plt.show()
```
